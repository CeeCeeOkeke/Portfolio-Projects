Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv.zip')
```

```python
# loading the first 5 rows and columns of our data
credit_card_data.head()
```

```python
# loading the last 5 rows and columns
credit_card_data.tail()
```

```python
#getting some information about the dataset
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction and fradulent transaction
credit_card_data['Class'].value_counts()
```

This dataset is Highly Unbalanced
1. 0 --------- Normal Transaction
2. 1 ----------- fraudlent transaction

```python
fraud = credit_card_data[credit_card_data.Class == 1]
legit = credit_card_data[credit_card_data.Class == 0]
```

```python
fraud.shape
```

```python
legit.shape
```

```python
# checking the Statistical measures of our data
credit_card_data.describe()
```

```python
# comparing the values for both transactions
credit_card_data.groupby('Class').mean()
```

Under Sampling

Building a sample dataset containing similar distribution of Normal transaction and Fraudulent transaction

Number of Fraudulent Transactions ----------------> 492

```python
legit_sample = credit_card_data.sample(n=492)
```

Concatenating the two Data Frame

```python
new_dataset = pd.concat([legit_sample, fraud],  axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.value_counts('Class')
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Target

```python
X = new_dataset.drop('Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
print(Y)
```

Test Splitting the Data

```python
X_test,X_train,Y_test,Y_train = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X,Y)
```

```python
training_model = model.predict(X_train)
```

```python
accuracy_training_model = accuracy_score(Y_train, training_model)
```

```python
print(f'Accuracy on training data is : {accuracy_training_model}')
```

```python
# Testing the model
testing_data = model.predict(X_test)
```

```python
accuracy_testing_model = accuracy_score(Y_test, testing_data)
```

```python
print(f'Accuracy on testing data is : {accuracy_testing_model}')
```
