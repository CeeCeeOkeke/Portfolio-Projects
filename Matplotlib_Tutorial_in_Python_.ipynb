Importing MatplotLibrary

```python
import matplotlib.pyplot as plt
```

```python
#Importing numpy to get data
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Ploting the data

```python
#Sine wave
plt.plot(x,y)
plt.show
```

```python
#Ploting the Cosinre graph
print(plt.plot(x,z))

```

```python
# Adding title to x-axis and y-axis
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel('sine value')
plt.title('sine wave')
plt.show
```

```python
#ploting a parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x,y)
plt.title('Parabola Graph')
plt.show()
```

```python
#printing in red colour
plt.plot(x, y, 'r*')
plt.show()
```

```python
#Prints in green colour
plt.plot(x, y, 'g*')
plt.show
```

```python
#Ploting two different graphs together differentiating them with differnt colours
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r+')
plt.title('SINE & COSINE GRAPH')
plt.show()

```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0, 1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0, 1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color= 'g')
ax.scatter(x,z, color='b')
plt.title('SINE AND COSINE IN SCATTER FORM')
plt.show()
```

Scatter plot in 3D

```python
fig3 = plt.figure()
ax = plt.axes(projection= '3d' )
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z, cmap='Blues')
plt.show()
```
